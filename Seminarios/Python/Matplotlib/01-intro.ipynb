Al graficar con MatPlotLib se utilizan se debe tener en cuenta distintos aspectos que hacen a la presentacioń del gráfico, en la imagen marca con círculos las zonas a las que hace referencia:
* Figure (fig) todo el espacio de la grafica.
* Axes, (axs) cada grafico.
* yAxis y xAxis, (yaxis, xaxis) los ejes "y" y "x".
* Legend (ax.legend) leyenda del grafico.
* ...
Estos elementos son editables al momento de definir el gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np
l1=np.arange(1,11)
l2=np.random.randint(0,10,10)
```

Existen 2 metodos para graficar:
* PyPlot
    * plt.plot(datosX,datosY)
    * plt.xlabel('Leyenda del eje X')...
* OOP
    * fig, ax = plt.subplots(figsize=(ancho,alto))
    * ax.plot(datosX,datosY)
    * ax.set_label(...)

* Grafico superpuesto de lineas con plot.

```python
plt.figure(figsize=(10,5))
plt.plot(l1,l2,label='serie1')
plt.plot(l1,l3,label='serie2')
plt.plot(l1,l4,label='serie3')
plt.xlabel('valores de X')
plt.ylabel('Valores de Y')
plt.title('Ejemplo')
plt.legend()
plt.show()
```

### OOP

```python
fig, ax=plt.subplots(figsize=(10,5))
ax.plot(l1,l2,label='serie1')
ax.plot(l1,l3,label='serie2')
ax.plot(l1,l4,label='serie')
ax.set_xlabel('valores de x')
ax.set_ylabel('valores de y')
ax.set_title('Ejemplo')
ax.legend()
plt.show
```

## Tipos de lineas

https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
* linewidth: Ancho
* linestyle: Estilo (puede ir la palabra o el símbolo)
    * dotted (:) : de puntos
    * dashed (--) : lineas
    * dashdot (-.) : punto y lineas

```python
plt.figure(figsize=(10,5))
plt.plot(l1,l2,label='serie1',linewidth=3,linestyle=':')
plt.plot(l1,l3,label='serie2',linewidth=4,linestyle='--')
plt.plot(l1,l4,label='serie3',linewidth=6,linestyle='-.')
plt.xlabel('valores de X')
plt.ylabel('Valores de Y')
plt.title('Ejemplo')
plt.legend()
plt.show()
```

* Parametrizando el trazo, *linestyle(x,(a,b,c,d))* comienza a dibujar la linea con desplazamiento x (generalmente comienza en 0) y luego intercambia entre trazo y espacio en blanco.

```python
plt.figure(figsize=(10,5))
plt.plot(l1,l2,label='serie1',linewidth=3,linestyle=(0,(4,2,10,1)))
plt.stem(l1,l2,label='serie3',markerfmt='d')
plt.show()
```
