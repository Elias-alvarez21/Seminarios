## Leer Datos de Archivos
Se pueden acceder a gran cantidad de tipos de archivos principalmente se utilizan archivos csv y se leecon *pd.read_x* donde *x* generalmente es el tipo de archivo, en este caso utilizaremos csv.<br>
Al imprimirlos, por defecto muestra las primeras y las ultimas filas del *Data Frame*<br>
Para mostrar todo el contenido a la variables debe agregarse *.to_strig()*

```python
import pandas as pd
```

```python
salarios = pd.read_csv('./indice_salarios.csv')
salarios
```

Otro formato muy tulizado para compartir tablas son los archivos de excel, para leerlos se utiliza pd.read_excel("archivo.xlsx")

Podemos convertir formatos de archivos para lograr compatibilidad con otros sistemas a travez de la función *.to_(formato)()* en este caso se convierte el un xlsx en scv, colocando en index todos los registros e incluyendo los nombres de las cabeceras:

```python
salxls = pd.read_excel('./Salarios.xlsx')
salxls.to_csv('./Salarios.csv', index=None, header=True)
```

## Reemplazar los datos null o faltantes con otro dato
* Se utiliza *.fillna(´dato´)* para reemplazar con *'dato'* el campo faltante

```python
salarios = pd.read_csv('./indice_salarios.csv')
salarios
```

```python
salarios.fillna(200,inplace=True)
salarios
```

* Para aplicar la corrección solo a una columna se debe indexar al momento de aplicar el filtro

```python
salarios = pd.read_csv('./indice_salarios.csv')
salarios["IS_indice_total"].fillna(200,inplace=True)
salarios
```

* Puede resultar ser necesario colocar un valor acorde al común de los valores de la columna, por lo tanto podemos colocar la media como valor, para esto se calcula la media de la columna en una variable y se coloca esa variable en el filtro de corrección.<br>
* Esto nos permite tener distontos valores para cada columna.

```python
salarios = pd.read_csv('./indice_salarios.csv')
sector=salarios["IS_sector_no_registrado"].mean()
total=salarios["IS_indice_total"].mean()
salarios["IS_sector_no_registrado"].fillna(sector.round(2),inplace=True)
salarios["IS_indice_total"].fillna(total.round(2),inplace=True)
print(salarios.to_string())
```
