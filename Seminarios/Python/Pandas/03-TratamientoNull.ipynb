Realizar un listado con los totales de vehículos Diesel por año

```python
for a in [2023,2024,2025]:
    print(a)
    print(ve.loc[(ve["year"] == a)&(ve["fuel"]=="Diesel")].count()["name"])
```

```python
ve[ve["fuel"]=="Diesel"].groupby('year').count()['name']
```
