```python
pelis=pd.read_csv('./Netflix.csv')
pelis
```

```python
rt=pd.read_csv('./Rating.csv')
rt
```

* Al dataframe de peliculas combinarle por id_rate, los nombres de las categorías.

```python
jo=pelis.join(rt,on=('id_rating'))
jo
```

* Eliminar las columnas con valores NaN.

```python
pelis.isna().any()
```

* Eliminar las columnas Description y id_rating.

```python
jo.drop(['director','cast','country','date_added','duration','description','id_rating'], axis=1, inplace=True)
jo
```

* Cual es la categoría mas frecuente de Film.

```python
jo['listed_in'].mode()
```

* Listar la cantidad de películas por año de lanzamiento.

```python
jo.groupby('release_year').count()
```

* Lista de los titulos con Categoría NaN.

```python
jo[jo['rating'].isna()]
```

* Listar la cantidad de películas de cada Categoría.

```python
jo.groupby('listed_in').count().sort_values('show_id', ascending=False)
```

* Listar de las 20 temáticas mas populares entre todas los registros.

```python
jo.groupby('listed_in').count().sort_values('title', ascending=False).head(20)
```

* Listar los años que tuvieron entre 20 y 30 lanzamientos de solo Drama o solo Kids' TV ordenado por año.

```python
dryk=jo.loc[(jo['listed_in']=='Drama')|(jo['listed_in']=='Kids\' TV')].groupby('release_year').count()
dryk.loc[(dryk['title'] > 20)&(dryk['title'] < 30)].sort_index()
```

* Realizar una tabla que compare las temáticas "Drama", "Comedia" y "Documentales" comparando la cantidad de lanzamientos en los últimos 5 años.

```python
drama=jo[jo['listed_in']=='Dramas'].groupby('release_year').count()[-5:]['show_id']
docs=jo[jo['listed_in']=='Documentaries'].groupby('release_year').count()[-5:]['show_id']
coms=jo[jo['listed_in']=='Comedies'].groupby('release_year').count()[-5:]['show_id']
graf=pd.DataFrame(data=[drama,docs,coms], index=["Dramas","Documentales","Comedias"])
graf
```

* Realizar un gráfico de barras comparativo del punto anterior, agregar Título, etiquetas de los ejes valores en las columnas, leyenda superior derecha y grilla horizontal con lineas de puntos.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
g=graf.T
ancho=0.25
print(g)
x=np.arange(len(g.index))
print(x)
fig, ax=plt.subplots()

dr=ax.bar(x-ancho,g.Dramas,ancho,color="#995555",label='Dramas')
do=ax.bar(x,g.Documentales,ancho,color="#559955",label='Documentales')
co=ax.bar(x+ancho,g.Comedias,ancho,color="#555599",label='Comedias')
ax.set_ylim(0,100)
ax.set_title('Comparativa de tipo de peliculas por año')
ax.set_ylabel('Cantidad de lanzamientos')
ax.set_xlabel('Años')
ax.legend(loc='upper right')
ax.set_xticks(x,g.index)
ax.bar_label(dr)
ax.bar_label(do)
ax.bar_label(co)

ax.grid(axis='y', linestyle=':')
plt.show()
```

* Graficar comparativa entre los tipos de contenido desde 2015, marcar el punto maximo de TV Show con una etiqueta, sin spins superior ni derecho, marcar cada serie con distintos marcadores, tipo de linea y colores.

```python
u=jo[jo["release_year"]>2014]
u
```

```python
ul=u.pivot_table(index='release_year', columns='type', values='title', aggfunc='count')
ul
```

```python
fig,ax=plt.subplots(figsize=(15, 5))
ax.plot(ul.index, ul.Movie,'rs:', ul.index,ul['TV Show'],'bd--')
ax.legend(['Películas','Shows'],loc='lower center')
ax.set_title('Peliculas / Shows desde 2015')
ax.set_xlabel('Años')
ax.set_ylabel('Lanzamientos')
ax.spines[['top','right']].set_visible(False)
ax.annotate('Maximo Shows',(2020,430),xytext=(2020,250), fontsize=13, arrowprops=dict(arrowstyle="-|>", facecolor='red'))
plt.show()
```
