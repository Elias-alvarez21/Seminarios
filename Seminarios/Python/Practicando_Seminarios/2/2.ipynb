```python
import pandas as pd
tita=pd.read_csv("titanic.csv")
tita
```

```python
p148=tita.query("PassengerId == 148")
print(p148)
filas_pares = tita.iloc[::2]
print(filas_pares)
```

```python
priClass=tita.query("Pclass == 1").sort_values("Name")
priClass
```

```python
porViv=tita[tita.Survived == 1].groupby(by="Survived")["Name"].count().reset_index()
porMuer=tita[tita.Survived == 0].groupby(by="Survived")["Name"].count().reset_index()
sum=porMuer+porViv

calmuer=(porMuer/sum)*100
print(calmuer)
calViv=(porViv/sum)*100
print(calViv)
print(porViv)
```

```python

PrimClass=tita.query("Pclass == 1")
PrimClass.rename(columns={"Name":"Cantidad"}, inplace=True)
vivPrimClass=PrimClass[(PrimClass.Survived == 1) | (PrimClass.Survived == 0)].groupby(by="Survived")["Cantidad"].count().reset_index()
total1=vivPrimClass.Cantidad.sum()
por1=(136/total1)*100
print("Porcentaje de vivos en primera clase",por1)

SegClass=tita.query("Pclass == 2")
SegClass.rename(columns={"Name":"Cantidad"}, inplace=True)
vivSegClass=SegClass[(SegClass.Survived == 1) | (SegClass.Survived == 0)].groupby(by="Survived")["Cantidad"].count().reset_index()
total2=vivSegClass.Cantidad.sum()
por2=(87/total2)*100
print("Porcentaje de vivos en segunda clase",por2)

TerClass=tita.query("Pclass == 3")
TerClass.rename(columns={"Name":"Cantidad"}, inplace=True)
vivTerClass=TerClass[(TerClass.Survived == 1) | (TerClass.Survived == 0)].groupby(by="Survived")["Cantidad"].count().reset_index()
total3=vivTerClass.Cantidad.sum()
por3=(119/total3)*100
print("Porcentaje de vivos en tercera clase",por3)
```

```python
tita.dropna(subset=["Age"],inplace=True)
tita
```
