```python
eje1=df[(df["Valor Medio de"] == "Temperatura (°C)")].min().reset_index()
#eje1=df.loc[:,[ 'Valor Medio de', 'Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
#       'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']].groupby(by="Estación")
eje1
```

```python
eje2=df.loc[(df["Valor Medio de"]=="Humedad relativa (%)") | (df["Valor Medio de"]=="Precipitación (mm)") | (df["Valor Medio de"]=="Temperatura (°C)")].dropna()
print(eje2)
eje2.isnull().any()
```

```python
grafPlata=eje2.query("Estación == 'LA PLATA AERO'")
grafPlata=grafPlata.set_index("Valor Medio de")
#grafPlataMES=grafPlata.drop(columns=["Estación","Valor Medio de"])
#grafPlataMES.transpose().reset_index()
grafPlata.drop(columns=["Estación"],inplace=True)
grafPlata=grafPlata.transpose().reset_index()


grafIguazu=eje2.query("Estación == 'IGUAZÚ AERO'")
grafIguazu=grafIguazu.set_index("Valor Medio de")
grafIguazu.drop(columns=["Estación"],inplace=True)
grafIguazu=grafIguazu.transpose().reset_index()

grafMendoza=eje2.query("Estación == 'MENDOZA AERO'")
grafMendoza=grafMendoza.set_index("Valor Medio de")
grafMendoza.drop(columns=["Estación"],inplace=True)
grafMendoza=grafMendoza.transpose().reset_index()

grafNeuquen=eje2.query("Estación == 'NEUQUÉN AERO'")
grafNeuquen=grafNeuquen.set_index("Valor Medio de")
grafNeuquen.drop(columns=["Estación"],inplace=True)
grafNeuquen=grafNeuquen.transpose().reset_index()
```

```python
import matplotlib.pyplot as plt

plt.subplot(221)
plt.ylim(0,250)
plt.title("LA PLATA")
plt.plot(grafPlata["index"],grafPlata["Temperatura (°C)"], color="blue",linestyle='-.',label='Temp')
plt.plot(grafPlata["index"],grafPlata["Humedad relativa (%)"], color="green", linestyle=':',label='Humedad')
plt.plot(grafPlata["index"],grafPlata["Precipitación (mm)"], color="red",label='Lluvia')
plt.xticks(rotation=90)
plt.subplot(222)
plt.ylim(0,250)
plt.title("IGUAZÚ")
plt.plot(grafIguazu["index"],grafIguazu["Temperatura (°C)"], color="blue",linestyle='-.',label='Temp')
plt.plot(grafIguazu["index"],grafIguazu["Humedad relativa (%)"], color="green", linestyle=':',label='Humedad')
plt.plot(grafIguazu["index"],grafIguazu["Precipitación (mm)"], color="red",label='Lluvia')
plt.xticks(rotation=90)
plt.subplot(223)
plt.ylim(0,250)
plt.title("MENDOZA")
plt.plot(grafMendoza["index"],grafMendoza["Temperatura (°C)"], color="blue",linestyle='-.',label='Temp')
plt.plot(grafMendoza["index"],grafMendoza["Humedad relativa (%)"], color="green", linestyle=':',label='Humedad')
plt.plot(grafMendoza["index"],grafMendoza["Precipitación (mm)"], color="red",label='Lluvia')
plt.xticks(rotation=90)
plt.subplot(224)
plt.ylim(0,250)
plt.title("NEUQUEN")
plt.plot(grafNeuquen["index"],grafNeuquen["Temperatura (°C)"], color="blue",linestyle='-.',label='Temp')
plt.plot(grafNeuquen["index"],grafNeuquen["Humedad relativa (%)"], color="green", linestyle=':',label='Humedad')
plt.plot(grafNeuquen["index"],grafNeuquen["Precipitación (mm)"], color="red",label='Lluvia')
plt.legend(loc='upper right')
plt.xticks(rotation=90)
plt.show()
```
